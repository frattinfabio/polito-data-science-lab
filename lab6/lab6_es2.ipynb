```python
import csv
import matplotlib.pyplot as plt
import numpy as np

X = []
labels = []
with open("syntetic2d.csv") as f:
    next(f)
    for cols in csv.reader(f):
        X.append([float(cols[0]), float(cols[1]), (float(cols[0]) + float(cols[1]))])
        labels.append(int(cols[2]))

X = np.array(X)    
fig,ax = plt.subplots(figsize=(10,10))
ax.scatter(X[:,0],X[:,2], c = labels)
plt.show()
        
```

```python
from sklearn.tree import DecisionTreeClassifier
import pydot
from sklearn.tree import export_graphviz

feature_names = ["x", "y", "x + y"]
clf = DecisionTreeClassifier().fit(X,labels)

dot_code = export_graphviz(clf, feature_names = feature_names, rounded=True)
with open("dot_code.txt", "w") as f:
    f.write(dot_code)
```

```python
tresh = clf.tree_.threshold[0]

fig,ax = plt.subplots(figsize=(12,12))
ax.scatter(X[:,0],X[:,2], c = labels)
plt.axhline(y = tresh, linewidth=1, color='r')
plt.show()
```

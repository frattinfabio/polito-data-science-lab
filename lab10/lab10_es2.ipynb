```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("831394006_T_ONTIME.csv", parse_dates = ["FL_DATE", "CRS_DEP_TIME","DEP_TIME","CRS_ARR_TIME","ARR_TIME"], infer_datetime_format = True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
num_un_car = df["UNIQUE_CARRIER"].nunique()
print(f"There are {num_un_car} unique carriers")
num_un_air = df["ORIGIN"].append(df["DEST"]).nunique()
print(f"There are {num_un_air} unique airports")
```

```python
# filtering out cancelled fligths
df = df[df['CANCELLED'] == 0]
```

```python
# counting how many flights per each carrier
df.groupby("UNIQUE_CARRIER")["FL_NUM"].agg('count').to_frame("TOT_FLIGHTS")
```

```python
# computing the average delay
df["TOT_DELAY"] = df["CARRIER_DELAY"]+df["WEATHER_DELAY"]+df["NAS_DELAY"]+df["SECURITY_DELAY"]+df["LATE_AIRCRAFT_DELAY"]
df.groupby("UNIQUE_CARRIER")["TOT_DELAY"].mean().to_frame("AVG_DELAY")
```

```python
# adding two new columns
df["DELAYDELTA"] = df["ARR_DELAY"]-df["DEP_DELAY"]
df["DAYOFWEEK"]= df["FL_DATE"].dt.dayofweek
```

```python
# looking for correlation between dow and arrival delay
arr_delay_dow = df.groupby("DAYOFWEEK")['ARR_DELAY'].mean().to_frame("AVG_ARR_DELAY")
fig,ax = plt.subplots(figsize=(12,6))
arr_delay_dow.plot.barh(ax = ax)
```

```python
# comparing arrival delay during weeknend and working days for each carrirer

weekend_df = df[np.logical_or(df["DAYOFWEEK"] == 5, df["DAYOFWEEK"] == 6)]
avg_weekend_arrival_delay = weekend_df.groupby("UNIQUE_CARRIER")["ARR_DELAY"].mean()

working_df = df[np.logical_and(df["DAYOFWEEK"] != 5, df["DAYOFWEEK"] != 6)]
avg_working_arrival_delay = working_df.groupby("UNIQUE_CARRIER")["ARR_DELAY"].mean()

carriers = sorted(df["UNIQUE_CARRIER"].unique())

map_df = {"AVG_WEEKEND_ARR_DELAY":avg_weekend_arrival_delay, "AVG_WORKING_ARR_DELAY": avg_working_arrival_delay }
arr_delay_df = pd.DataFrame(map_df)

fig,ax = plt.subplots(figsize=(15,9))
arr_delay_df.plot.barh(ax = ax)
```

```python
#  multi indexing over unique_carrier, origin, dest, fl_date
indexed_df = df.set_index(keys = ['UNIQUE_CARRIER','ORIGIN','DEST','FL_DATE'])
```

```python
#  For each ﬂight operated by American Airlines (AA) and Delta Airlines (DL), taken oﬀ from the Los Angeles International Airport (LAX) and for each date, display the departure time and delay. 
indexed_df.loc[['AA','DL'],'LAX',:,:][["FL_NUM","DEP_TIME","DEP_DELAY"]]
```

```python
# For each ﬂight that ﬂew in the ﬁrst week of the month, with LAX as destination airport, compute the mean arrival delay. 
df[np.logical_and(df["DEST"]=='LAX', df["FL_DATE"].dt.day <=7)]["ARR_DELAY"].mean()
```

```python
# Generate a pivot table containing the number of departed ﬂights for each carrier and for each day of the week and show it. 
pivot_df = df.pivot_table(index = "DAYOFWEEK", columns = "UNIQUE_CARRIER", aggfunc='count', values = "FL_NUM")
corr_mat = pivot_df.corr()
fig, ax = plt.subplots(figsize=(12,12))

sns.heatmap(corr_mat, annot=True, cmap="Greens")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
# Generate a pivot table containing the average arrival delay, for each carrier and for each day of the week and show it. 
pivot_df = df.pivot_table(index = "DAYOFWEEK", columns = "UNIQUE_CARRIER", values = "ARR_DELAY", aggfunc='mean')

corr_mat = pivot_df.corr()
fig, ax = plt.subplots(figsize=(12,12))

sns.heatmap(corr_mat, annot=True, cmap="Greens")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
#  Using a pivot table, for the carriers HA, DL, AA and AS compute the average deltadelay for each day of the week. Then, display the results on a line plot, having a line per carrier and the weekday on the x-axis.
pivot_df = df.pivot_table(index = "DAYOFWEEK", columns = "UNIQUE_CARRIER", values = "DELAYDELTA", aggfunc='mean')
ix = pd.IndexSlice
pivot_df = pivot_df.loc[:, ix["HA","DL","AA","AS"]]
fig,ax = plt.subplots(figsize=(15,10))
pivot_df.plot.line(ax = ax)
```

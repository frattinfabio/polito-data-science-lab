```python
# LAB 3 es.1

# loading the dataset into a dictionary: each column is a (key,list of values) pair
import csv
import numpy as np

col_names = [] 
# InvoiceNo,StockCode,Description,Quantity,InvoiceDate,UnitPrice,CustomerID,Country
dataset = [[], [], [], [], [], [], [], []]

with open("online_retail.csv") as f:
    reader = csv.reader(f)
    col_names = next(reader)
    
    for cols in reader:
        if len(cols) == len(col_names):
            if(not cols[0].startswith('C')):
                for i in range(len(col_names)):
                    dataset[i].append(cols[i])
```

```python
aggregated_lists = {}
for i in range (len(dataset[0])):
    invoice_no = dataset[0][i]
    if(aggregated_lists.get(invoice_no, 0) == 0):
        aggregated_lists[invoice_no] = []
    aggregated_lists[invoice_no].append(dataset[2][i]) 
```

```python
distinct_items = list(set(dataset[2]))
matrix = []
for key, value in aggregated_lists.items():
    matrix.append([(item in value) for item in distinct_items])
    
```

```python
import pandas as pd

df = pd.DataFrame(data = matrix, columns = distinct_items)
```

```python
from mlxtend.frequent_patterns import fpgrowth

fi = fpgrowth(df, 0.01)
print(fi.to_string())
```

```python
from mlxtend.frequent_patterns import association_rules

ass_rules = association_rules(fi, metric='confidence', min_threshold=0.85)

for i in range(len(ass_rules['antecedents'])):
    print(f"{i + 1})", end = " ")
    for j in range(len(ass_rules['antecedents'][i])):
        if j == len(ass_rules['antecedents'][i]) - 1:
            end = " "
        else:
            end = ", "
        print(distinct_items[list(ass_rules['antecedents'][i])[j]], end = end)
    print("=>", end = " ")
    for j in range(len(ass_rules['consequents'][i])):
        if j == len(ass_rules['consequents'][i]) - 1:
            end = " "
        else:
            end = ", "
        print(distinct_items[list(ass_rules['consequents'][i])[j]], end = end)
    print("")    
```
